```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import cv2 
import os
```

```python
img = plt.imread("sample-1.jpg")
```

```python
plt.figure(figsize = (12,10))
plt.imshow(img)
plt.colorbar()
```

```python
plt.figure(figsize = (12, 10))
img_gray = cv2.imread("sample-1.jpg")
gray = cv2.cvtColor(img_gray, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap = "gray")
plt.colorbar()
plt.title("Gray image of original image")
```

```python
plt.figure(figsize = (8, 6))
plt.hist(gray.flatten(), bins=256, range=[0, 256], color = "violet")
plt.legend(["Histogram"], loc = "upper right")
plt.show()
```

```python
plt.figure(figsize=(10, 8))
a = plt.subplot(1, 2, 1)
a.set_title("Original Image")
plt.imshow(gray, cmap = "gray")

b = plt.subplot(1, 2, 2)
plt.hist(gray.flatten(), bins=256, range=[0, 256], color = "violet")
plt.legend(["Hist"], loc = "upper right")
b.set_title("Original Image Histogram")
plt.show()
```
